### Dropping out Irrelevant Coulmns

```python
data.rename(columns={'iyear':'Year','imonth':'Month','iday':"Day",'gname':'Group','country_txt':'Country',
                     'region_txt':'Region','provstate':'Province','city':'City','latitude':'Latitude',
                     'longitude':'Longitude','summary':'Summary','attacktype1_txt':'Attacktype',
                     'targtype1_txt':'Targettype','weaptype1_txt':'Weapon','nkill':'Killed',
                     'nwound':'Wounded'},inplace = True)
```

```python
data = data[['Year','Month','Day','Country','Province','Region','City','Latitude','Longitude',"Attacktype",
             'Killed','Wounded','target1','Summary','Group','Targettype','Weapon','motive']]
data.head()
```

### Visualizing The Data

#### Heatmap

```python
plt.figure(figsize=(10,10))
sns.heatmap(data.corr(), annot= True)
plt.show()
```

### Timeline: Terrorist Activities by Years

```python
plt.subplots(figsize=(20,10))
sns.countplot('Year',data=data,palette='cubehelix',edgecolor=sns.color_palette("dark:salmon_r", 10))
plt.xticks(rotation=50)
plt.title('Terrorist Activities Each Year',fontsize=24)
plt.show()
```

### Weapons Most Used By Terrorists

```python
data.Weapon.value_counts()[:10]
```

```python
plt.subplots(figsize=(20,10))
sns.barplot(data['Weapon'].value_counts()[:10].index, data['Weapon'].value_counts()[:10].values, palette='magma')
plt.title("Top 10 Weapons Used in Terror Attacks", fontsize=24)
plt.xlabel('Weapon Type', fontsize=18)
plt.ylabel('No. Of Weapon Used', fontsize=18)
plt.xticks(rotation = 90)

plt.show()
```

### Top 10 Countries with Highest Total number of Terror Attacks

```python
data.Country.value_counts()[:10]
```

```python
plt.subplots(figsize=(20,10))
country = data['Country'].unique()
countries_count = data['Country'].value_counts(dropna = False).sort_index()
sns.barplot(x=country, y=countries_count)
sns.barplot(data['Country'].value_counts()[:10].index,data['Country'].value_counts()[:10].values,palette='rocket_r')
plt.title("Countries with Highest Number of Total Attacks", fontsize=24)
plt.xlabel('Countries', fontsize=18)
plt.ylabel('No. of Attacks', fontsize=18)
plt.xticks(rotation=90)
plt.show()
```

### Top 10 Cities targetted by Terrorists

### Timeline of Frequency of Attacks over the world

```python
plt.subplots(figsize=(12,7))
sns.kdeplot(data['Year'], hue=data['Region'])
plt.title('Region wise Terrorist Activities', fontsize=24)
plt.xlabel('Year', fontsize=18)
plt.ylabel('Frequency Of Attacks', fontsize=18)
plt.xticks(rotation = 90)
plt.show()
```

### Pie Chart

```python
plt.subplots(figsize=(20,10))
data.groupby(['Year'])['Killed'].sum().sort_values(ascending = False).head(20).plot(kind = 'pie')
plt.title("Years with Highest Number of deaths due to terrorist activities", fontsize=24)
plt.xlabel('Attack Year', fontsize=18)
plt.ylabel('No. Of Deaths', fontsize=18)
plt.xticks(rotation = 90)
plt.show()
```

#### Total Number of Deaths by Terrorism

```python
killData = data.loc[:,'Killed']
print('Number of people killed by terror attack:', int(sum(killData.dropna())))

countryData = data.loc[:,'Country']
countryKillData = pd.concat([countryData, killData], axis=1)
```

#### Country Wise Number of Deaths by Terrorism 

```python
countryKillFormatData = countryKillData.pivot_table(columns='Country', values='Killed', aggfunc='sum')
countryKillFormatData
```

## Plotting Country wise Total Deaths Due to Terrorist Activities

```python
fig_size = plt.rcParams["figure.figsize"]
fig_size[0]=25
fig_size[1]=25
plt.rcParams["figure.figsize"] = fig_size
```

```python
labels = countryKillFormatData.columns.tolist()
labels = labels[:50] #50 bar provides nice view
index = np.arange(len(labels))
transpoze = countryKillFormatData.T
values = transpoze.values.tolist()
values = values[:50]
values = [int(i[0]) for i in values] # convert float 
colors = sns.color_palette("viridis")
# color list for bar chart bar color 
fig, ax = plt.subplots(1, 1)
ax.yaxis.grid(True)
fig_size = plt.rcParams["figure.figsize"]
fig_size[0]=25
fig_size[1]=25
plt.rcParams["figure.figsize"] = fig_size
plt.bar(index, values, color = colors, width = 0.9)
plt.ylabel('Killed People', fontsize=30)
plt.xlabel('Countries', fontsize = 30)
plt.xticks(index, labels, fontsize=20, rotation=90)
plt.title('Number of People Killed In Countries', fontsize = 30)
# print(fig_size)
plt.show()
```
